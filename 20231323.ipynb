# 1. LeetCode 225. Implement Stack using Queues


```python
# LeetCode 225 풀이 코드 (Improved)
from collections import deque

class MyStack:
    def __init__(self):
        self.q = deque()

    def push(self, x: int) -> None:
        self.q.append(x)
        for _ in range(len(self.q) - 1):
            self.q.append(self.q.popleft())

    def pop(self) -> int:
        return self.q.popleft()

    def top(self) -> int:
        return self.q[0]

    def empty(self) -> bool:
        return not self.q
```

# 2. LeetCode 232. Implement Queue using Stacks


```python
# LeetCode 232 풀이 코드
class MyQueue:

    def __init__(self):
        self.in_stack = []
        self.out_stack = []

    def push(self, x: int) -> None:
        self.in_stack.append(x)

    def pop(self) -> int:
        self.peek()
        return self.out_stack.pop()

    def peek(self) -> int:
        if not self.out_stack:
            while self.in_stack:
                self.out_stack.append(self.in_stack.pop())
        return self.out_stack[-1]

    def empty(self) -> bool:
        return not self.in_stack and not self.out_stack
```

## 01

```python
class ListQueue:
    def __init__(self):
        self.__queue = []

    def enqueue(self, x):
        # 리스트의 맨 앞에 추가
        self.__queue.insert(0, x)

    def dequeue(self):
        # 리스트의 맨 뒤에서 제거
        return self.__queue.pop()

    def front(self):
        # 리스트의 맨 뒤가 큐의 front
        return self.__queue[-1]

    def isEmpty(self) -> bool:
        return len(self.__queue) == 0

    def dequeueAll(self):
        self.__queue.clear()
```

## 02

```python
from collections import deque

def is_palindrome(s):
    q = deque(s)
    while len(q) > 1:
        if q.popleft() != q.pop():
            return False
    return True

# 테스트
print(is_palindrome("wow"))  # True
print(is_palindrome("word"))  # False

```

## 03

```python
# a에서 b로 복사 (단순 복사 아님, 내용만 복사)

class LinkedQueue:
    def __init__(self):
        self.data = []

    def enqueue(self, x):
        self.data.append(x)

    def dequeue(self):
        return self.data.pop(0)

    def is_empty(self):
        return len(self.data) == 0

    def front(self):
        return self.data[0]

a = LinkedQueue()
b = LinkedQueue()

for val in [1, 2, 3]:
    a.enqueue(val)

temp = LinkedQueue()

while not a.is_empty():
    x = a.dequeue()
    b.enqueue(x)
    temp.enqueue(x)

while not temp.is_empty():
    a.enqueue(temp.dequeue())

print(a.data)  # [1, 2, 3]
print(b.data)  # [1, 2, 3]
```

## 04

```python
class StackWithQueues:
    def __init__(self):
        self.q1 = []
        self.q2 = []

    def push(self, x):
        self.q1.append(x)

    def pop(self):
        while len(self.q1) > 1:
            self.q2.append(self.q1.pop(0))
        popped = self.q1.pop(0)
        self.q1, self.q2 = self.q2, []
        return popped
```

## 05

```python
class QueueUsingStacks:
    def __init__(self):
        self.stack1 = []
        self.stack2 = []

    def enqueue(self, x):
        self.stack1.append(x)

    def dequeue(self):
        if not self.stack2:
            while self.stack1:
                self.stack2.append(self.stack1.pop())

        if self.stack2:
            return self.stack2.pop()
        else:
            return None  # 큐가 비어 있음
```

## 06

 enqueue()는 FIFO 구조이므로 입력크기에 상관없이 항상 일정한 시간 안에 실행된다. 따라서  수행시간은 O(1)이다

 dequeue()도 FIFO 구조로 언제나 빠르게 실행가능하며 수행시간은 O(1)로 동일하다

## 07

 enqueue()의 수행시간은 O(n)이며 맨 끝까지 탐색이 필요한 반면 dequeue()의 수행시간은 O(1)로 더미 헤드 덕분에 바로 삭제 가능하다.

```python
class ListDeque:
    def __init__(self):
        self.__deque = []

    def enqueueFront(self, x):
        self.__deque.insert(0, x)  # 맨 앞에 삽입 (시간: O(n))

    def enqueueRear(self, x):
        self.__deque.append(x)     # 맨 뒤에 삽입 (시간: O(1))

    def dequeueFront(self):
        if not self.isEmpty():
            return self.__deque.pop(0)  # 맨 앞 제거 (시간: O(n))
        return None

    def dequeueRear(self):
        if not self.isEmpty():
            return self.__deque.pop()   # 맨 뒤 제거 (시간: O(1))
        return None

    def front(self):
        if not self.isEmpty():
            return self.__deque[0]
        return None

    def rear(self):
        if not self.isEmpty():
            return self.__deque[-1]
        return None

    def isEmpty(self):
        return len(self.__deque) == 0

    def clear(self):
        self.__deque.clear()

    def printDeque(self):
        print("Deque from front:", end=" ")
        for item in self.__deque:
            print(item, end=" ")
        print()
```
